```
import random
import matplotlib.pyplot as plt
```

```
#batasan range
x_range = 30
y_range = 30

ax = plt.axes(xlim=(0, x_range), ylim=(0, y_range)) 
line, = ax.plot([], []) 
```

```
#inisiasi partikel

n_particle = 10
x_pos = []
y_pos = []

for i in range(n_particle):
    x_pos.append(random.randint(0, 20))
    y_pos.append(random.randint(0, 30))
```

```
#update posisi dan cek boundary
n_iter = 100
```

```
x_pos2 = [[0 for i in range(n_iter)] for j in range(n_particle)]
y_pos2 = [[0 for i in range(n_iter)] for j in range(n_particle)]
rand = random.uniform(0, 1)
x = 0
y = 0
old_x = 0
old_y = 0
for i in range(n_particle):
  for j in range(n_iter):
    if j == 0:
        x_pos2[i][j] = x_pos[i]
        y_pos2[i][j] = y_pos[i]
        old_x = x_pos[i]
        old_y = y_pos[i]
    else:
      rand = random.uniform(0, 1)
      if rand <=  0.25:   #kanan 2x Kiri
        x += 1
      elif rand <= 0.125: #bawah 2x atas
        y -= 1
      elif rand <= 0.125: #kiri 1/2 kanan
        x -= 1
      else:               #atas 2x kiri
        y += 1
      newX = x + old_x 
      newY = y + old_y
      if newX > x_range:
            newX = 0
      elif newX < 0:
            newX = x_range
      if newY > y_range:
            newY = 0
      elif newY < 0:
          newY = y_range   
      x_pos2[i][j] = newX
      y_pos2[i][j] = newY 
      old_x = newX
      old_y = newY
      x = 0
      y = 0
   
```

```
for i in range(n_particle):
  plt.plot(x_pos2[i], y_pos2[i])
plt.show()
```
